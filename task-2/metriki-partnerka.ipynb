### Метрики мобильного приложения
Основные метрики которые будем использовать:

1. DAU, MAU - оцениваем степень активности в приложении;
2. Retention Rate (RR) - оцениваем вовлеченность и заинтересованность людей (в разрезе каналов коммуникаций). Далее смотрим на аудиторию и сегментируем ее, выделяя группы с минимальной и максимальной конверсией в целевое событие (транзакция у партнера) и одновременно следим за тем, откуда приходят самые платежеспособные клиенты.
3. Session - время, которое пользователь провел в приложении;
4. ASV (Average Session Duration) - среднее время сессий за период времени;
5. Есть еще одна очень интересная метрика и неплохо было бы ее использовать: ATV (average transaction value) - средняя стоимость покупок за период (или средний чек). Данная метрика позволит сегментировать пользователей на сегменты с “маленьким”, “средним” и “большим” средним чеком. Потом можно сравнивать эти данные о активности. И уже на основе этих данных строить маркетинговую стратегию, к примеру. По сути это получается некий аналог классического RFM-анализа.
6. Одной из необязательных, но вполне досягаемых метрик на мой взгляд является частное от DAU к MAU (т.е. DAU/MAU). Так называемый Stickness Rate - грубо говоря, это показатель “залипательности” приложения. Однако, если оценивать данную метрику в динамике, то можно увидеть ее изменчивость и понять где мы больше проседаем и активность каких пользователей надо увеличивать.

Очень классно было бы построить поэтапную воронку событий (установка, запуск, онбоардинг, транзакции, RR): начиная сверху с момента установки (будем считать это первым моментом взаимодействия с продуктом) и в глубь вплоть до RR. Воронка позволит увидеть, что происходит с нашими пользователями на каждом этапе и кто когда начинает усиленно отваливаться. Для чего это нужно: а) чтобы понять на каком этапе улучшать (где мы теряем больше всего) и б) для того чтобы понять какой потенциальный профит будет от изменений

### Набор систем для аналитики

*Базовая аналитика*

Для базовой аналитики на начальном этапе можно обойтись установкой в приложение одной из имеющихся систем аналитики и разметкой событий, которые планируется отслеживать. Например: AppMetrica, Firebase, Amplitude, AppsFlyer.

На данном этапе необходимо будет сделать пошагово следующие вещи:

1. Определиться с системой трекинга данных;
2. Подготовить ТЗ для установки SDK аналитической системы для разработчиков приложения (если еще не установлена);
3. Подготовить карту событий для разметки в приложении;
4. Внедрить аналитику в приложение;
5. Протестировать сбор данных.

*Расширенная аналитика*

Для более расширенного варианта анализа пользователей приложения необходимо эти данные начать обогащать или добавлять в систему новые данные. Усилить аналитику можно будет за счет создания DWH, в которую будут собираться данные о действиях пользователей из разных источников (например, рекламные системы, системы трекинга данных в приложении, данные о пользователях и другие).

Все эти вещи необходимо в том числе и для того чтобы, построить полноценную UJM. 

*User journey, или UJM* — это карта пути пользователя внутри продукта. То есть непосредственно того человека, который будет использовать продукт. Создать UJM можно, в частности, с использование Firebase и Google Analytics. Одним из других инструментов предоставления целостной и подробной атрибуции является также *AppsFlyer*. Вообще в части построения User Journey мне кажется AppsFlyer нет равных: развитая модель атрибуции, технология диплинкинга и возможность управления проведением А/В тестов - делает платформу незаменимой для аналитики и мобильного маркетинга.

### Разметка приложения

Процесс разметки in-app событий и подготовки технического задания для разработчиков прежде всего связан с разметкой основных событий в воронке. Здесь можно выделить основные следующие этапы: 

1. Начало сессии;
2. Открытие каталога (партнерской витрины);
3. Просмотр страницы товара;
4. Добавление товара в корзину;
5. Оформление заказа (совершение транзакции).

Это начальный, преаналитический, этап. Далее на втором этапе происходит определение событий, которые являются уникальными для каждого отдельного интерфейса и приложения. Другими словами, где эти элементы расположены, на каких экранах, этапах воронки, связаны ли элементы между собой какой-то логикой или механикой. Второй этап является более скрупулезным, потому что затрагивает каждый элемент интерфейса приложения, с которым пользователь может взаимодействовать и от результата взаимодействия с которым зависит дальнейшее поведение пользователя.
